```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' , names=['sepal length' , 'sepal width' , 'petal length' , 'petal width', 'class'])

# Adding column names to a read dataset, of names dont exist
```

```python
df.head()
```

```python
df_setosa = df.loc[df['class']=='Iris-setosa']
df_versicolour = df.loc[df['class']=='Iris-versicolor']
df_virginica = df.loc[df['class']=='Iris-virginica']
```

```python
print(df_setosa.shape)
print(df_versicolour.shape)
print(df_virginica.shape)
```

# Univariate Analysis

### Using Sepal Length

```python
plt.plot(df_setosa['sepal length'], np.zeros_like(df_setosa['sepal length']),'o')
plt.plot(df_versicolour['sepal length'], np.zeros_like(df_versicolour['sepal length']),'o')
plt.plot(df_virginica['sepal length'], np.zeros_like(df_virginica['sepal length']),'o')
plt.show()
```

### Using Sepal Width

```python
plt.plot(df_setosa['sepal width'], np.zeros_like(df_setosa['sepal width']),'o')
plt.plot(df_versicolour['sepal width'], np.zeros_like(df_versicolour['sepal width']),'o')
plt.plot(df_virginica['sepal width'], np.zeros_like(df_virginica['sepal width']),'o')
plt.show()
```

### Using Petal Length

```python
plt.plot(df_setosa['petal length'], np.zeros_like(df_setosa['petal length']),'o')
plt.plot(df_versicolour['petal length'], np.zeros_like(df_versicolour['petal length']),'o')
plt.plot(df_virginica['petal length'], np.zeros_like(df_virginica['petal length']),'o')
plt.show()
```

### Using Petal Width

```python
plt.plot(df_setosa['petal width'], np.zeros_like(df_setosa['petal width']),'o')
plt.plot(df_versicolour['petal width'], np.zeros_like(df_versicolour['petal width']),'o')
plt.plot(df_virginica['petal width'], np.zeros_like(df_virginica['petal width']),'o')
plt.show()
```

# Bivariate Analysis

### Using Petal Length vs Petal Width

```python
sns.FacetGrid(df, hue='class', height=4).map(plt.scatter, 'petal length', 'petal width').add_legend()
plt.show()
```

### Using Sepal Length vs Sepal Width

```python
sns.FacetGrid(df, hue='class', height=4).map(plt.scatter, 'sepal length', 'sepal width').add_legend()
plt.show()
```

# Multivariate Analysis

```python
sns.pairplot(df,hue='class',height=3)
plt.show()
```
